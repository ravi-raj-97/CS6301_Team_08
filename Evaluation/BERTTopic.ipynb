```python
#Function to remove stopwords from the Summary to get more accurate topics
def remove_stopwords(doc):
  doc=doc.split()
  doc=[x for x in doc if x.lower() not in stop_words]
  doc=' '.join(doc)
  lines=doc.split('.')
  return lines
```

```python
#Get a list of possible topics for the give summary
def get_topics(topic_model):
  df=topic_model.get_topic_info().head(5)
  df['Name']=df['Name'].apply(lambda x:'_'.join(x.split('_')[1:]))
  return str(dict(zip(df.Topic,df.Name)))
```

```python
#Function to return topics genereated when passed a Text
def model_topic(doc):
  topic_model = BERTopic()
  lines=remove_stopwords(doc)
  if(len(lines)<11):
    return None
  topics, probs = topic_model.fit_transform(lines)
  #print(str(topic_model.get_topic_info()))
  result=get_topics(topic_model)
  return result
```
