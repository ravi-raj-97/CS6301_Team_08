```python
# Mount the drive if running on colab 
from google.colab import drive
drive.mount('/content/drive')
# Import other required libraries 
import pandas as pd
from summarizer import Summarizer
from tqdm.notebook import tqdm_notebook
import tqdm
tqdm.tqdm.pandas()
import warnings
warnings.filterwarnings('ignore')
```

```python
# Read the file that contains the data that needs to be summarized
data_to_test = pd.read_excel('/content/drive/MyDrive/summary_3_and_25_complete.xlsx')
data_to_test.head(2)
```

```python
# Load the summarizer
model = Summarizer()
```

```python
# Function to generate summary of a transcript
# summary size/transcript size is the ratio_val
def get_sum(transcript, ratio_val):
  small_transcript = model(transcript, ratio=ratio_val)
  return small_transcript
```

```python
# get summary of transcript
# replace 'json_transcript' with column name that has the transcript
# replace 'summary_35' with the column in which the data needs to be stored
data_to_test['summary_35'] = data_to_test['json_transcript'].progress_apply(lambda x: get_sum(x,0.35))
```

```python
# write the data to file
data_to_test.to_excel('/content/drive/MyDrive/summary_35_generated.xlsx', index=False)
```
